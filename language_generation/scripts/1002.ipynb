```python
def plot_performance_via_uncertainty(result_1, result_2, num_bins = 10, metric = 'bert_scores'):
    x = []
    y = []
    for i in range(len(result_1['valid_loss'])):
        x.append(result_2['valid_loss'][i])
        y.append(compare(result_1['valid_loss'][i], result_2['valid_loss'][i]))
    plt.figure(figsize=(20, 6))
    plt.cla()
    x=np.array(x)
    y=np.array(y)
    
    sorted_x = sorted(x)  # 对数据进行排序，得到索引
    group_size = len(x) // num_bins  # 每个分组的数据数量
    remainder = len(x) % num_bins  # 余数，用于平衡每组的数据数量
    bin_sizes = [group_size + 1 if i < remainder else group_size for i in range(num_bins)]  # 每个分组的数据数量列表
    bin_edges = np.array(sorted_x)[[np.sum(bin_sizes[:i+1]) - 1 for i in range(num_bins)]]
    bin_edges = np.insert(bin_edges,0,0)
    bin_edges = np.unique(bin_edges)
    num_bins = len(bin_edges) - 1  # 计算每个分组内y数据的平均值
    bin_y1_avg = [np.mean(y[(x >= bin_edges[i]) & (x < bin_edges[i + 1])]) for i in range(num_bins)]
    
    plt.xticks(range(num_bins), [f'{"%.1f"%(bin_edges[i])}-{"%.1f"%(bin_edges[i+1])}({len(y[(x >= bin_edges[i]) & (x < bin_edges[i + 1])])})' for i in range(num_bins)])
    plt.plot(range(0,num_bins), bin_y1_avg,)
    plt.xlabel('Certainty score',fontsize=26)
    plt.ylabel(f'2v2 accuracy',fontsize=26)
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=26)
    plt.legend()
    plt.show()
    plt.cla()
    x = []
    y1 = []
    y2 = []
    for i in range(len(result_1['valid_loss'])):
        x.append(result_2['valid_loss'][i])
        y1.append(result_2['valid_loss'][i])
        y2.append(result_1['valid_loss'][i])
    plt.scatter(x,y1)
    plt.scatter(x,y2)
    plt.show()
    
result_1, result_2 = result['gpt2-xl'][0], result['gpt2-xl'][1] 
result_1, result_2 = result['gpt2-medium'][0], result['gpt2-medium'][1]
# result_1, result_2 = result['llama-7b'][0], result['llama-7b'][1]
plot_performance_via_uncertainty(result_1, result_2)    
    
```
